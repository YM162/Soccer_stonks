```python
#Hack estrellas infinitas
import requests
import random
import string
from time import sleep
from random import randint
```

```python
def randomString(stringLength=8):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(stringLength))
```
